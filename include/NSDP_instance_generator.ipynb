```julia
function create_NFSs(dp::Int64,cp::Int64,folder::String,instance_number::Int64,lat::String)
    
    open(joinpath(folder,"instance_$(instance_number)_NFS_types.dat"), "w") do io
        write(io, "number_of_VNFs $(cp+dp)\n")
        write(io, "number_of_DP_NFs $(dp)\n")
        write(io, "number_of_CP_NFs $(cp)\n\n")
            
        for vnf in 1:(dp)
            write(io, "NFS $(vnf) id $(vnf) type dp treatment_capacity $(rand(15.0:0.001:30.0)) dataVolPerUE 0.001 cpu_request $(rand(1:2)).0 ram_request $(rand(8:12)).0  storage_request $(rand(800:1200)).0  compression 1.0 UE_dependency 1 amount_of_traffic_sent_to_g ")
            for vnf in 1:(dp+cp)
                 write(io, "0.001 ")
            end  
            write(io, "\n")
        end
            
        write(io, "\n")
            
        for vnf in dp+1:(dp+cp)
            write(io, "NFS $(vnf) id $(vnf) type cp treatment_capacity $(rand(0.5*5.5:0.001:5.5)) dataVolPerUE 0.001 cpu_request $(rand(1:2)).0 ram_request $(rand(8:12)).0  storage_request $(rand(800:1200)).0  compression 1.0 UE_dependency 1 amount_of_traffic_sent_to_g ")
            for vnf in 1:(dp+cp)
                 write(io, "0.000001 ")
            end
            write(io, "\n")
        end

    end #end of VNF_instances
    
    my_matrix_connexion = zeros(Int8,(cp+dp, cp+dp))
    my_matrix_latency = zeros(Float64,(cp+dp, cp+dp))
    for f in 1:(cp+dp-1),g in (f+1):(cp+dp)
    
        if f <=dp && g<= dp && f == g-1
            my_matrix_connexion[f,g] = 1
            my_matrix_connexion[g,f] = 1
        
        elseif  g > dp
            randd = rand(0:1)
            my_matrix_connexion[f,g] = randd
            my_matrix_connexion[g,f] = randd
        end
        if lat == "low"
            randd = rand(0.5:0.1:1.5)
        else
            randd = rand(2.0:0.1:4.0)
        end
        my_matrix_latency[f,g] = randd
        my_matrix_latency[g,f] = randd
    end 
        
    open(joinpath(folder,"instance_$(instance_number)_NFSs_to_connect.dat"), "w") do io
        open(joinpath(folder,"instance_$(instance_number)_latency_btw_functions.dat"), "w") do ioo
            write(io,"number_of_NFSs $(dp+cp)\n\n") 
            write(ioo,"number_of_NFSs $(dp+cp)\n\n")
            for f in 1:(cp+dp)
                for g in 1:(cp+dp)
                    write(io,"$(my_matrix_connexion[f,g]) ")
                    write(ioo,"$(my_matrix_latency[f,g]) ")
                end
                write(io,"\n")
                write(ioo,"\n")
            end
        end
    end
    
end
```

```julia
function create_physical_topology(instance_number::Int64,n::Int64,d::Float64,c::String,folder::String,number_nfss::Int64,k::Int64)
    #creating basic graph with n nodes and average degree equals to d
    g = 0
    connected = false
    while connected ==  false 
        g = SimpleGraph(n)
        for u in 1:n-1, v in u+1:n
            if rand(0:0.0001:1) <= d
                add_edge!(g, u, v)
            end
            
        end
        connected = is_connected(g)    
    end
    
    
    
    count_arc_id = 1    
    my_nodes = 1:n

    dus = Vector{Int64}()
    apps = Vector{Int64}()
    open(joinpath(folder,"instance_$(instance_number)_physical_network.dat"), "w") do io
        write(io, "number_of_nodes $(n)\n")
        write(io, "number_of_arcs $(2*length(edges(g)))\n")
        write(io, "node_capacity_types 4 storage_capacity networking_capacity ram_capacity cpu_capacity\n\n")          
       
            #creating access nodes
            while length(dus)<trunc(Int,n/3)
                new_du = rand(1:n)
                if  (new_du ∈ dus) == false
                    write(io, "node $(new_du) node_id $(new_du) node_type access longitude $(rand(10:1000)/1.023) latitude $(rand(0:1000)/1.023) coverage_radius $(rand(100:1000)/1.03) ")
                    if c == "tight"
                        write(io, "ram_capacity $(12*number_nfss).0 cpu_capacity $(2*number_nfss).0 storage_capacity  $(1200*number_nfss).0 networking_capacity  $(rand(4:4)).0 ")
                    else
                        write(io, "ram_capacity $(12*number_nfss*5).0 cpu_capacity $(2*number_nfss*5).0 storage_capacity  $(1200*number_nfss*5).0 networking_capacity  $(rand(4:4)).0 ")
                    end
                    write(io, "ram_cost $(rand(0.10:0.20)) cpu_cost $(rand(1.0:0.1:2.0)) storage_cost $(rand(0.01:0.015)) networking_cost $(rand(1:2)) ue_capacity $(rand(100:200)) availability 0.999 internal_delay 54 node_cost $(rand(100:200))\n\n")
                    push!(dus,new_du)
                    my_nodes = filter(x -> x != new_du, my_nodes)
                end
            end

            #creating application nodes
            while length(apps)<trunc(Int,n/3)
                new_app = rand(1:n)
                if  (new_app ∈ apps) == false && (new_app ∈ dus) == false 
                    write(io, "node $(new_app) node_id $(new_app) node_type app longitude $(rand(10:1000)/1.023) latitude $(rand(0:1000)/1.023) coverage_radius $(rand(100:1000)/1.03) ")
                    write(io, "ram_capacity 0 cpu_capacity 0 storage_capacity  0 networking_capacity  0 ")
                    write(io, "ram_cost 0 cpu_cost 0 storage_cost 0 networking_cost 0 bue_capacity 0 availability 0.999 internal_delay 54 node_cost 0\n\n")
                    push!(apps,new_app)
                    my_nodes = filter(x -> x != new_app, my_nodes)
                end
            end


            #creating non-access nodes
            for i in my_nodes
                write(io, "node $(i) node_id $(i) node_type non_access longitude $(rand(10:1000)/1.023) latitude $(rand(0:1000)/1.023) coverage_radius $(rand(100:1000)/1.03) ")
                if c == "tight"
                    write(io, "ram_capacity $(12*number_nfss).0 cpu_capacity $(2*number_nfss).0 storage_capacity  $(1200*number_nfss).0 networking_capacity  $(rand(4:4)).0 ")
                else
                    write(io, "ram_capacity $(12*number_nfss*5).0 cpu_capacity $(2*number_nfss*5).0 storage_capacity  $(1200*number_nfss*5).0 networking_capacity  $(rand(4:4)).0 ")
                end
                write(io, "ram_cost $(rand(0.10:0.20)) cpu_cost $(rand(1.0:0.1:2.0)) storage_cost $(rand(0.01:0.015)) networking_cost $(rand(1:2)) ue_capacity $(rand(100:200)) availability 0.999 internal_delay 54 node_cost $(rand(100:200))\n\n")
            end

            
            for a in edges(g)
                latency = rand(0.25:0.01:1.0)
                if c == "tight"
                    bandwidth = rand(30*0.5*k:0.1:30*k)
                else
                    bandwidth = rand(30*2*k:0.1:30*3*k)
                end
                write(io, "arc $(count_arc_id) edge_id $(count_arc_id) source $(src(a)) target $(dst(a)) type fiber delay $(latency) max_bandwidth $(bandwidth) availability 0.999 cost 0.1\n\n")
                count_arc_id=count_arc_id+1
                write(io, "arc $(count_arc_id) edge_id $(count_arc_id) source $(dst(a)) target $(src(a)) type fiber delay  $(latency)  max_bandwidth $(bandwidth) availability 0.999 cost 0.1\n\n")
                count_arc_id=count_arc_id+1
            end
        end# closing io file
    return dus,apps
                                                          
end
```

```julia
function create_commodities(du::Array{Int64,1}, app::Array{Int64,1}, folder::String, number_commodities::Int64, number_slices::Int64,number_of_instances::Int64)
    for s in 1:number_slices
        open(joinpath(folder,"instance_$(number_of_instances)_commodities_slice$(s).dat"), "w") do io
            write(io, "number_of_commodities $(trunc(Int,number_commodities))\n\n")
            aux_du = du
            volume = rand(10.0:0.1:50.0)
            for com in 1:number_commodities
                 if length(aux_du) > 0
                     rand_du = rand(aux_du)
                     write(io, "commodity_id $(trunc(Int,com)) origin_node $(trunc(Int,rand_du)) target_node $(rand(app)) volume_of_data $(volume)\n")
                     aux_du = filter(x -> x != rand_du, aux_du)
                else
                    break
                end
            end
        end
    end
end
```

```julia
function create_NS_requests(folder::String,number_slices::Int64,instance_number::Int64,lat::String)
    open(joinpath(folder,"instance_$(instance_number).dat"), "w") do io
       quo = '"'
       write(io, "number_of_requests $(number_slices)\n\n")
        for s in 1:number_slices
            write(io, "slice_request $(s) id $(s) ")
            write(io, "total_amount_of_UE $(rand(1000:10000)) min_bandwidth $(rand(5.0:0.1:10.0)) ")
            write(io, "VNF_types_to_install  $(quo)$(folder)instance_$(instance_number)_NFSs_to_install.dat$(quo) ")
            write(io, "VNFs_to_connect  $(quo)$(folder)instance_$(instance_number)_NFSs_to_connect.dat$(quo) ")
            write(io, "VNF_sharing  $(quo)$(folder)instance_$(instance_number)_NFS_sharing_slice$(s).dat$(quo) ")
            write(io, "commodities  $(quo)$(folder)instance_$(instance_number)_commodities_slice$(s).dat$(quo) ")
            if lat == "low"
                write(io, "maxLatencyDataLayer $(rand(2.5:0.1:5.0)) ")
            else  
                 write(io, "maxLatencyDataLayer $(rand(3.0:0.1:10.0)) ")
            end
            write(io, "nodeSharing  $(quo)$(folder)instance_$(instance_number)_node_sharing.dat$(quo)\n\n")     
        end
        
    end
end
```

```julia
function create_node_sharing(folder::String, number_slices::Int64,number_of_instances::Int64,node_sharing::Float64)
    open(joinpath(folder,"instance_$(number_of_instances)_node_sharing.dat"), "w") do io
        for s in 1:number_slices
            for t in 1:number_slices
                rand_sharing = rand(0.0:0.01:1.0)    
                if rand_sharing <= node_sharing && node_sharing>0.0000001
                    write(io,"1 ")
                else
                    write(io,"0 ")       

                end  
            end
            write(io,"\n")
        end
    end
end
```

```julia
function create_NFS_sharing(number_of_NFSs::Int64,folder::String, number_slices::Int64,number_of_instances::Int64,nfs_sharing::Float64)
    for s in 1:number_slices
        open(joinpath(folder,"instance_$(number_of_instances)_NFS_sharing_slice$(s).dat"), "w") do io
            write(io,"number_of_VNFs $(number_of_NFSs)\n") 
            write(io,"number_of_slices $(number_slices)\n\n")
            for t in 1:number_slices
                for f in 1:number_of_NFSs
                    for g in 1:number_of_NFSs
                        rand_sharing = rand(0.0:0.00001:1.0)    
                        if rand_sharing <= nfs_sharing && nfs_sharing>0.0000001
                            write(io,"1 ")
                        else
                            write(io,"0 ")       
                        end 
                    end
                    write(io,"\n")
                end
                 write(io,";\n")
            end
        end
    end
end        
```

```julia
function create_NFS_to_install(folder::String, number_slices::Int64,number_of_instances::Int64,dp::Int64,cp::Int64)
    open(joinpath(folder,"instance_$(number_of_instances)_NFSs_to_install.dat"), "w") do io
        quo = '"'
        write(io,"set_VNF_to_install $(quo)instance_$(number_of_instances)$(quo)\n")
        write(io,"number_of_VNFs_to_install $(dp+cp)\n\n")
        
        for s in 1:number_slices
            for f in 1:dp+cp
                if f <= dp+1
                    write(io,"1 ")
                else
                    write(io,"$(rand(0:1)) ")       

                end  
            end
            write(io,"\n")
        end
    end
end
```
